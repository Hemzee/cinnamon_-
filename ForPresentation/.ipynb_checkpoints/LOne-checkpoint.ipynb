### Start to make Trainning Data

```python
training_data = []
```

```python
def create_training_data():
    
    low_brown = (8,100,20)
    high_brown = (16,255,200)
    aspect_ratio = 1.33
    img_width = 252
    img_height = math.ceil(img_width/aspect_ratio)         #Add ceil to make rounded numbers 3.6 -> 4
    
    for category in CATEGORIES:                            #["D1","D2","D3"]
        
        path = os.path.join(DATADIR,category)
        
        class_num = CATEGORIES.index(category)
        
        for img in os.listdir(path):
            try:
                img = cv2.imread(os.path.join(path,img))
                hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
                brown_mask = cv2.inRange(hsv,low_brown,high_brown)
                brown = cv2.bitwise_and(img,img,mask = brown_mask)
                
                new_array = cv2.resize(brown,(img_width, img_height))
                training_data.append([new_array, class_num])                #array and classNum  0,1,2
            except Exception as e: 
                pass
            
        print(category + " Set : ",np.array(training_data).shape)
        
        
```

```python
create_training_data()
```

```python
for a,b in training_data:
    print(a," :: ",b)
```

```python
import random
random.shuffle(training_data)    #to shuffle the data set
```

```python
#creating two axises for feeding

x_axis=[]
y_axis=[]
```

```python
for features,label in training_data:         #features= arrays, labels = [0,1,2]
    x_axis.append(features)
    y_axis.append(label)
        
    
X_axis_Set = np.array(x_axis).reshape(-1, img_width , img_height, 3)   
#X_axis_Set

```

### Save created X and Y axises to files for reuse.

```python
pickle_out = open("X_axis","wb")
pickle.dump(X_axis_Set, pickle_out)
pickle_out.close()


pickle_out = open("y_axis","wb")
pickle.dump(y_axis, pickle_out)
pickle_out.close()
```
